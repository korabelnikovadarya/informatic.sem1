**Задача 7. (2 балла)** Создайте с помощью генератора списков (`list comprehension`) список из чисел от -10 до 10. С помощью функции `filter` удалите из него положительные числа, кратные 3. Затем с помощью функции `map` возведите оставшиеся числа в квадрат. Выведите результат на экран.

```python
from random import randint
a = [randint(-10, 10) for i in range(10)]
def f(x):
    return x % 3 == 0
def ff(x):
    return x * x
a = list(filter(f,a))
a = list(map(ff,a))
print(a)
```

**Задача 9. (3 балла + 1 балл)** Вводится натуральное число n. Выведите все пифагоровы тройки с целочисленными катетами, в которых длина гипотенузы не превосходит n. Формат вывода: в столбец, в каждой строке по 3 числа, первые 2 из которых - катеты. Выводить можно в любом порядке. Для получения четвертого балла решение должно быть в одну строку без `;`.

Пример:

*Входные данные:*

`13`

*Выходные данные:*

`3 4 5`

`4 3 5`

`6 8 10`

`8 6 10`

`5 12 13`

`12 5 13`

```python
a = int(input())
for j in range(1, a):
    for i in range(j + 1, a + 1):
        b = int((i * i - j * j)**0.5)
        if b**2 + j*j == i*i:
            print(b, j, i)
```

**Задача 10. (4 балла)**

"В условиях предыдущей задачи" - это про формат данных: фамилия, имя, школа, балл

```python
n = int(input())
maxx = 0
sum_all = 0
a = [[0,0] for i in range(n + 1)]
for i in range(n):
    b = list(map(str, input().split()))
    a[i][0] = int(b[2])
    a[i][1] = int(b[3])
    sum_all += int(b[3])
    if int(b[2]) + int(b[3]) > maxx:
        maxx = int(b[2]) + int(b[3])
a[n][0] = maxx
a[n][1] = maxx # добавляем последний элемент в список а, чтобы
# считывание происходило до последней школы включительно
a.sort(key = lambda x: x[0]) # сортируем по номеру школы

first = a[0][0]
s = a[0][1]
summ = []
numbschool = []
count = []
coun = 1
# суммируем баллы по школам
for i in range(1, n + 1):
    if a[i][0] == first:
        s += a[i][1]
        coun += 1
    else:
        summ.append(s)
        numbschool.append(first)
        count.append(coun)
        s = 0
        coun = 0
        first = a[i][0]
sr = sum_all / n # средний по всем школам

# summ -  сумма баллов по школам по отдельности
summ = [summ[i] / count[i] for i in range(len(count))]
for i in range(len(count)):
    if summ[i] > sr:
            print(numbschool[i])

```
